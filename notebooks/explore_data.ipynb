```python
import nltk
nltk.download('movie_reviews')
from nltk.corpus import movie_reviews

print(f"Antall anmeldelser: {len(movie_reviews.fileids())}")
print(f"Kategorier: {movie_reviews.categories()}")
```

```python
# Hent filene
pos_filer = movie_reviews.fileids(categories='pos')
neg_filer = movie_reviews.fileids(categories='neg')

print(f"Positive anmeldelser: {len(pos_filer)}")
print(f"Negative anmeldelser: {len(neg_filer)}")

# Se på en faktisk anmeldelse
en_positiv = movie_reviews.raw(pos_filer[0])
print("\n--- POSITIV ANMELDELSE ---")
print(en_positiv[:500])
```

```python
en_positiv = movie_reviews.raw(pos_filer[0])
print("--- POSITIV ANMELDELSE ---")
print(en_positiv[:500])

en_negativ = movie_reviews.raw(neg_filer[0])
print("\n--- NEGATIV ANMELDELSE ---")
print(en_negativ[:500])
```

```python

ord = movie_reviews.words(pos_filer[0])
print(f"Antall ord i anmeldelsen: {len(ord)}")
print(f"De første 20 ordene: {list(ord[:20])}")
```
